```python
from typing import List
nums = [1, 1, 1, 2, 3,4,4,4,5,6,6,7,8]
len(nums)
```

```python
# 有点慢
pre = None
idx = 0
while idx < len(nums):
    cur = nums[idx]
    if cur == pre:
        nums.pop(idx)
    else:
        pre = cur
        idx += 1
print(len(nums))
print(nums)
```

```python
# 略快一些
nums = [1,1,2]
for idx in range(len(nums) - 1, 0, -1):
    if nums[idx] == nums[idx - 1]:
        nums.pop(idx)
```

```python
# 双指针
def removeDuplicates(self, nums: List[int]) -> int:
        if len(nums) < 2:
            return len(nums)
        l, r = 0, 1
        while r < len(nums):
            lv, rv = nums[l], nums[r]
            if rv != lv:
                l += 1
                nums[l] = rv
            r += 1
        return l + 1
```

```python
nums = [1, 1,2]

left, right = 0, 1
while right < len(nums):
    print(f" {'   ' * left}{left}")
    print(nums)
    print(f" {'   ' * right}{right}")   
    if nums[left] != nums[right]:
        left += 1
        nums[left] = nums[right]
    right += 1
nums = nums[:left + 1]
print(len(nums))
print(nums)
```
