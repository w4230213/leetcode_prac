```python
nums = [1, 1, 1, 2, 3,4,4,4,5,6,6,7,8]
len(nums)
```

```python
pre = None
idx = 0
while idx < len(nums):
    cur = nums[idx]
    if cur == pre:
        nums.pop(idx)
    else:
        pre = cur
        idx += 1
print(len(nums))
print(nums)
```

```python
for idx in range(len(nums) - 1, 0, -1):
    if nums[idx] == nums[idx - 1]:
        nums.pop(idx)
```

```python
nums = [1, 1,2]

left, right = 0, 1
while right < len(nums):
    print(f" {'   ' * left}{left}")
    print(nums)
    print(f" {'   ' * right}{right}")   
    if nums[left] != nums[right]:
        left += 1
        nums[left] = nums[right]
    right += 1
nums = nums[:left + 1]
print(len(nums))
print(nums)
```
