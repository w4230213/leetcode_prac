```python
p = [2,7,9,3,1]
```

```python
def dajiajieshe(p):
    if len(p) <= 2:
        return max(p)
    dp = [0 for _ in range(len(p))]
    dp[0] = p[0]
    dp[1] = max(p[:2])
    for i in range(2, len(p)):
        dp[i] = max(dp[i-2] + p[i],
                    dp[i-1])
    return max(dp)
```

```python
dajiajieshe(p)
```
