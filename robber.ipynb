```python
mv = {}
```

```python
# 递归解博物馆大盗

def thief(tr, m, mv):
    if tr == set() or m <= 0:
        mv[(tuple(tr),m)] = 0
        return 0
    elif (tuple(tr),m) in mv:
        return mv[(tuple(tr), m)]
    else:
        vmax = 0
        for t in tr:
            if t[0] <= m:
                v = thief(tr-{t}, m-t[0], mv) + t[1] # 假如一定要带上物品t, 这时的价值
                vmax = max(v, vmax)                  # 应该做一次判断
        mv[(tuple(tr),m)] = vmax
        return vmax
```

```python
print(thief(tr, max_w, mv))
```

```python
# 动态规划解博物馆大盗

tr = [None, (2,3), (3,4), (4,8), (5,8), (9,10), (11,20), (12,5)]
max_w = 30

v_matrix = {}
for i in range(len(tr)):
    for m in range(max_w + 1):
        if tr[i] == None or m == 0:
            v_matrix[(i,m)] = 0
        elif tr[i][0] > m:
            v_matrix[(i,m)] = v_matrix[(i-1,m)]
        else:
            # 假如第i件物品重量大,价值又低.那么为了装下i, 需要减少最大负重,可能导致之前(i-1)
            # 件物品中有的物品需要丢弃, 这可能得不偿失, 因此要做一次max()判断来取舍
            v_matrix[(i,m)] = max(v_matrix[(i-1,m-tr[i][0])] + tr[i][1], 
                                  v_matrix[(i-1), m])   
```
