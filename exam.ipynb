```python
def convert_format_2(input_list):
    out = {}
    rst = {}
    top_keys = set()
    for idx, item in enumerate(input_list):
        p, n = item.get('parent'), item['name']
        if not p:
            out.setdefault(n, {n: {}})
            top_keys.add(n)
        else:
            out.setdefault(p, {p: {}})
            out.setdefault(n, {n: {}})
            p_dict, n_dict = {p: {}}, {n: {}}
            out[p][p].update(out[n])
            # out[n].update(n_dict)
    for top_key in top_keys:
        rst.update(out[top_key])
    return rst
```

```python
%%timeit
rst = convert_format_2(input_list)
# print(json.dumps(rst, indent=4, ensure_ascii=False))
```

```python
def convert_format_1(input_list):
    out = {item['name']: {} for item in input_list if not item.get('parent')}
    rst = [{item['parent']: {item['name']: {}}} for item in input_list if item.get('parent')]
    par_list = [list(_item.keys())[0] for _item in rst]
    name_list = [list(list(_item.values())[0].keys())[0] for _item in rst]
    for idx, dd in enumerate(rst):
        _p, _n = par_list[idx], name_list[idx]
        if _p in out.keys():
            out[_p].update(dd[_p])
        if _n in par_list:
            child_idx_list = []
            for i, child in enumerate(par_list):
                if child == _n:
                    child_idx_list.append(i)
            for _i in child_idx_list:
                dd[_p][_n].update(rst[_i][_n])
    return out
```

```python
%%timeit
out = convert_format_1(input_list)
# print(json.dumps(out, indent=4, ensure_ascii=False))
```
