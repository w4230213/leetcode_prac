```python
digits = [9,9,9,9]
```

```python
def plusOne(digits):
    addone = True
    idx = 0
    _di = digits[::-1]
    while addone:
        tmp = _di[idx] + 1
        if tmp >= 10:
            _di[idx] = 0
            idx += 1
            if idx == len(digits):
                _di.append(1)
                addone = False
        else:
            _di[idx] = tmp
            addone = False
    return _di[::-1]
```
