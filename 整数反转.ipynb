```python
def reve(n):
    res = 0
    negative = n < 0
    if negative:
        n = - n
    while n > 0:
        tail = n % 10
        res = res * 10 + tail
        n //= 10
    res = -res if negative else res
    res = res if -2 << 30 < res < (2 << 30) - 1 else 0
    return res
```

```python
n = -54213

reve(n)
```

```python
def single(cls):
    _instance = {}
    def _single(*args, **kw):
        if cls not in _instance:
            _instance[cls] = cls(*args, **kw)
        return _instance[cls]
    return _single
```
