```python
def quicksort(nums, start, end):
    if start >= end:
        return
    i, j = start, end
    pivot = nums[i]
    while i < j:
        while i < j and nums[j] >= pivot:
            j -= 1
        nums[i], nums[j] = nums[j], nums[i]
#         nums[i][1], nums[j][1] = nums[j][1], nums[i][1]
        while i< j and nums[i] < pivot:
            i += 1
        nums[i], nums[j] = nums[j], nums[i]
#         nums[i], nums[j] = nums[j], nums[i]
    nums[j] = pivot
    quicksort(nums, start, j-1)
    quicksort(nums, j+1, end)
    return nums
```
