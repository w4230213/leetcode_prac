```python
def func(root, a, b):
    if not root or root is a or root is b:
        return root
    le = func(root.left, a, b)
    rt = func(root.right, a, b)
    if le and rt:
        return root
    if not le:
        return rt
    if not rt:
        return le
    return None
```

```python
def quicksort(nodes, start, end):
    if start >= end:
        return
    i, j = start, end
    pivot = nodes[i][0]
    while i < j:
        while i < j and nodes[j][0] >= pivot:
            j -= 1
        nodes[i][0], nodes[j][0] = nodes[j][0], nodes[i][0]
        nodes[i][1], nodes[j][1] = nodes[j][1], nodes[i][1]
        while i< j and nodes[i][0] < pivot:
            i += 1
        nodes[i][0], nodes[j][0] = nodes[j][0], nodes[i][0]
        nodes[i][1], nodes[j][1] = nodes[j][1], nodes[i][1]
    nodes[j][0] = pivot
    quicksort(nodes, start, j-1)
    quicksort(nodes, j+1, end)
    return nodes
```

```python
#node = (x, y)
def findk(nodes):
    max_k = 0
    flag_idx = 0
    sorted_nodes = quicksort(nodes, 0, len(nodes)-1)
    for i in range(len(sorted_nodes)-1):
#     for i in sorted_nodes):
        k = cal_k(sorted_nodes[i], sorted_nodes[i+1])
        if k > max_k:
            max_k = k
            flag_idx = i
    return sorted_nodes[flag_idx], sorted_nodes[flag_idx+1]
```

```python
def cal_k(node1, node2):
    return abs(node1[1] - node2[1]) / abs(node1[0] - node2[0])
```
