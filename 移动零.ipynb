```python
def movezero(nums):
    l, r = 0, 1
    while r < len(nums):
        if nums[l] == 0 and nums[r] == 0:
            r += 1
        elif nums[l] == 0 and nums[r] != 0:
            nums[l], nums[r] = nums[r], nums[l]
            r += 1
            l += 1
        else:
            l += 1
            r += 1
```

```python
# nums = [0,1,0,3,12, 0, 0 , 22, 0]
nums = [1,0]
movezero(nums)
print(nums)
```
