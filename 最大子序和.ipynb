```python
def maxsubsum(nums):
    if len(nums) == 1:
        return nums[0]
    cur = _max = nums[0]
    for i in range(1, len(nums)):
        cur = cur + nums[i] if cur >= 0 else nums[i]
        _max = max(_max, cur)
    return _max
```
