```python
prices = [7, 1, 88, 3, 6, 4]
# prices = [1,2,3,4,5]
# prices = [7,6,4,3,1]
```

```python
# pythonic
diff = [prices[r] - prices[r - 1] for r in range(1, len(prices))]

sum((i for i in diff if i >= 0))
```

```python
# DP
hold, unhold = -prices[0], 0
for i in prices[1:]:
    hold = max(hold, unhold - i)
    unhold = max(hold + i, unhold)
print(max(hold, unhold))
```
