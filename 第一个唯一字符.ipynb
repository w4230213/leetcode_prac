```python
def firtunique(st):
    count_dict = dict()
    for idx, char in enumerate(st):
        if count_dict.setdefault(char, [idx]) != [idx]:
            count_dict[char].append(idx)
#         if char not in count_dict:
#             count_dict[char] = [idx]
#         else:
#             count_dict[char].append(idx)
    tmp_list = [val[0] for val in count_dict.values() if len(val) == 1]
    if not tmp_list:
        return -1
    else:
        return min(tmp_list)
```

```python
def firstUniqChar(s):
    rst = -1
    for char in s:
        if s.find(char) == s.rfind(char):
            rst = s.find(char)
            break
    return rst
```
