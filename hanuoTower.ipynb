```python
def moveDisk(height, fromPole, toPole):
    print('move disk[{}] from {} to {}'.format(height, fromPole, toPole))

@clock
def moveTower(height, left, mid, right):
    if height >= 1:
        moveTower(height-1, left, right, mid)
        moveDisk(height, left, right)
        moveTower(height-1, mid, left, right)
```

```python
moveTower(2, 'left', 'mid', 'right')
```

```python
# 迷宫搜索,伪代码
def searchFrom(x, y):
    updatePos(x,y,S)
    if maze[x,y] == OBSTICLE:
        return False
    if maze[x,y] == TRID or DEADEND:
        return False
    if isExit(x,y):
        updatePos(x,y,PATH)
        return True
    updatePos(x,y,TRID)
    found = searchFrom(x-1,y) or \
            searchFrom(x+1,y) or \
            searchFrom(x, y-1) or \
            searchFrom(x, y+1)
    if found:
        updatePos(x, y, PATH)
    else:
        updatePos(x, y, DEADEND)
```
