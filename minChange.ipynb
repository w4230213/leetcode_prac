```python
# 递归法+缓存优化
#@clock
def recMC(coinValueList, change, knowResults):
    minCoin = change
    if change in coinValueList:
        knowResults[change] = 1
        return 1  
    elif knowResults[change] > 0:
        return knowResults[change]
    else:
        for i in [v for v in coinValueList if v <= change]:
            numCoin = recMC(coinValueList, change-i, knowResults) + 1
            if numCoin < minCoin:
                minCoin = numCoin
                knowResults[change] = minCoin
    return minCoin
```

```python
print(recMC([1,5,10,21,25], 63, [0]*64))
```

```python
# 动态规划解决找零问题    minCoinTable: 动态规划表
@clock
def dpMC(coinValueList, change, minCoinTable, usedCoin):
    for cents in range(1, change+1):
        minCoins = cents
        if cents in coinValueList:
            minCoinTable[cents] = 1
            usedCoin[cents] = cents
        else:
            for i in [v for v in coinValueList if v < cents]:
                numCoins = minCoinTable[cents-i] + 1
                if numCoins <= minCoins:
                    minCoins = numCoins
                    usedCoin[cents] = i
            minCoinTable[cents] = minCoins
    coinUsedList = []
    #print(usedCoin)
    # 用于输出使用了哪些硬币
    _change = change
    while _change > 0:
        thisCoin = usedCoin[_change]
        coinUsedList.append(thisCoin)
        _change -= thisCoin
    print('used coins --->', coinUsedList)
    return minCoinTable[change]
```

```python
coinValueList = [1,5,10,11,25]
change = 63
minCoinTable = [0] * (change+1)
usedCoin = dict()  # {c:0 for c in coinValueList}
print(dpMC(coinValueList, change, minCoinTable, usedCoin))
```
