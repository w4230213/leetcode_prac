```python
import turtle
from time import sleep

%matplotlib inline
```

```python
def luoxuan(length,t):
    if length > 0:
        t.forward(length)
        t.right(90)
        luoxuan(length-5,t)
```

```python
def tree(l,t):
    if l > 5:
        t.forward(l)
        sleep(0.5)
        t.right(20)
        t.pencolor('green')
        tree(l-15,t)
        t.left(40)
        t.pencolor('red')
        tree(l-15, t)
        t.right(20)
        t.pencolor('blue')
        t.backward(l)
```

```python
t = turtle.Turtle()
t.left(90)
t.penup()
t.backward(100)
t.pendown()
tree(75,t)
t.hideturtle()
turtle.done()
```
